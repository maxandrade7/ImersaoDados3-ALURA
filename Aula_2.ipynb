# **Aula 2 - Estatísticas, visualização de dados e distribuições**

```python
import pandas as pd
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true' 
dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True)
```

```python
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[:5]
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
```

#**Desafio 01: Ordenar o gráfico countplot**


```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos) #ordenando o gráfico countplot utilizando o paramêtro "order"
ax.set_title('Top 5 Compostos')
plt.show()
```


#**Desafio 02: Melhorar a visualização alterando tamanho da fonte...**

```python
sns.set()
plt.figure(figsize=(8,6))
ax = sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order=cod_compostos) #ordenando o gráfico countplot utilizando o paramêtro "order"
ax.set_title('Top 5 Compostos')
font = {'family':'sans-serif','color':'Black','size':18}
plt.title('Compostos', fontdict=font)
plt.xlabel('ID dos compostos',fontsize=13)
plt.ylabel('Quantidade', fontdict=font)
plt.show()
```

#**Desafio 03: Plotar os histogramas com seaborn**


```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dados, x="g-0", bins = 80)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dados, x="g-1", bins = 80)
plt.show()
```

#**Desafio 04: Estudar sobre as estatíticas retornadas no .describe()**


*count* é a quantidade de elementos

*mean* é a média

*std* é o desvio padrão

*min* é o valor mínimo

*25*% é o primeiro quartil

*50%* é a mediana

*75%* é o terceiro quartil

*max* é o valor máximo

#**Desafio 05: Refletir sobre a manipulação do tamanho das visualizações.**

A visualização de dados é uma parte essencial da análise de dados exploratórios. É mais eficaz do que números simples ao fornecer uma visão geral ou resumo de dados. As visualizações de dados nos ajudam a entender a estrutura subjacente dentro de um conjunto de dados ou explorar as relações entre variáveis. (Anello - Medium)

Portanto diferentes visualizações podem facilitar ou dificultar a mensagem a quem está observando os gráficos. Há diferentes gráficos e maneiras de fazer esta demonstração e isto pode fazer total diferença na compreensão.
